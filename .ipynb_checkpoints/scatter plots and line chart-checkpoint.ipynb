```python
df_control1=pd.read_csv('/Users/ananyahegde/Desktop/IEEE/data/control_new/control_1_new.csv')
```

```python
df_unique_date_condtn1=df_condtn1.date.unique()
unique_date_condtn1_week1=[]
for i in range(1,8):
    unique_date_condtn1_week1.append(df_unique_date_condtn1[i])

unique_date_condtn1_week1
```

```python
df_unique_date_control1=df_control1.date.unique()
unique_date_control1_week1=[]
for i in range(1,8):
    unique_date_control1_week1.append(df_unique_date_control1[i])

unique_date_control1_week1
```

```python
# df.reset_index(inplace=True)
df_app_condtn_week1 = pd.DataFrame()
for i in unique_date_condtn1_week1:
    selected_data = df_condtn1.loc[df_condtn1['date'] == i]
#     print(selected_data)
    df_app_condtn_week1 = pd.concat([df_app_condtn_week1,selected_data], axis=0)
# print(df_app_condtn.index)
df_app_condtn_week1.reset_index(inplace=True)
# print(df_app_condtn.index)
print(df_app_condtn_week1.shape)
```

```python
df_app_control_week1 = pd.DataFrame()
for i in unique_date_control1_week1:
    selected_data = df_control1.loc[df_control1['date'] == i]
#     print(selected_data)
    df_app_control_week1 = pd.concat([df_app_control_week1,selected_data], axis=0)
    
df_app_control_week1.reset_index(inplace=True)    
# print(df_app_control.index)
print(df_app_condtn_week1.shape)
```

```python
# plt.scatter(df_condtn1["hour"], df_condtn1["Avg_activity"], c='green')
```

```python
# plt.scatter(df_control1["hour"], df_control1["Avg_activity"], c='red')
```

```python
fig = plt.figure(figsize=(10,4))
ax1 = fig.add_subplot(111)
ax1.scatter(df_app_condtn_week1.index, df_app_condtn_week1["Avg_activity"],s=5, c='r', marker="s", label='condition')
ax1.scatter(df_app_control_week1.index, df_app_control_week1["Avg_activity"], s=5, c='g', marker="o", label='control')
plt.xlabel("hours")  
plt.ylabel("Avg_activity") 
plt.legend(loc='upper left')
plt.title("1_week scatter plot")
# plt.grid()
plt.show()
```

```python
fig = plt.figure(figsize=(8,4))
plt.plot(df_app_condtn_week1.index, df_app_condtn_week1['Avg_activity'], color='red', marker='o',label='condition')
plt.plot(df_app_control_week1.index, df_app_control_week1['Avg_activity'], color='green', marker='s',label='control')
plt.xlabel("hours")  
plt.ylabel("Avg_activity")  
plt.title("1_week line chart") 
plt.legend(loc='upper left')
plt.show()
# linestyle="-." - to change line syle
```
